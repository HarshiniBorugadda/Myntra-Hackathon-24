```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
data = data[["src", "alt"]]
data
```

```python
def parse_author(row):
  alt: str = row["alt"]
  start, end = alt.find(" by "), alt.find(" on ")
  author = alt[start+3:end].strip() if start and end else "Unknown"
  return author
```

```python
def parse_time(row):
  alt: str = row["alt"]
  start = alt.find(" on ")
  date = alt[start+3:start+17] if start else ""
  return date
```

```python
data["author"] = data.apply(parse_author, axis=1)
data
```

```python
data["time"] = data.apply(parse_time, axis=1)
data
```
